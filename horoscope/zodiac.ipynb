```python
import markovify
```

```python
text = open('text.txt', encoding='utf8').read().replace('\n', '. ')
text_model = markovify.Text(text)
```

```python
zodiac_signs = {
                1: "Овен",
                2: "Телец", 
                3: "Близнецы",
                4: "Рак",
                5: "Лев",
                6: "Дева",
                7: "Весы",
                8: "Скорпион",
                9: "Стрелец",
                10: "Козерог",
                11: "Водолей",
                12: "Рыбы"
                }

for num, sign in zodiac_signs.items():
    print(f'{num} -> {sign}')

zodiac = int(input("{blue}Введите число с номером знака зодиака: {endcolor}".format(blue="\033[96m", endcolor="\033[0m")))

if 0 < zodiac < 13:
    forecast = text_model.make_sentence()
    while not forecast:
        forecast = text_model.make_sentence()
    print(zodiac_signs[zodiac] + ', ' + forecast)
    
else:
    print("Вы ошиблись с числом, запустите программу ещё раз")
```
