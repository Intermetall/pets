```python
from matplotlib import pyplot as plt   
import random
```

```python
dice = 6
roll = 100000
```

```python
low = dice
high = dice * 6
result = [0] * (high + 1)
x = list(range(low, high))

for i in range(roll):
    summ = 0
    for i in range(dice):
        summ += random.randint(1, 6)
    result[summ] += 1


for i in range(dice+1):
    result.pop(0)
```

```python
plt.plot(x, result)   
plt.ylabel('Сколько раз выпало')   
plt.xlabel('Сумма бросков')   
plt.title(f'Распределение суммы бросков {dice} кубиков {roll} раз')
plt.show()
```
